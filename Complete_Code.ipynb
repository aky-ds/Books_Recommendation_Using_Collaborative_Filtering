```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
# Getting Ratings with everybook
ratings_with_df=ratings_with_name.groupby('Book-Title')['Book-Rating'].count().reset_index()
```

```python
# Calculating avg rating with books
avg_rating=ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
```

```python
avg_rating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
```

```python
popular=avg_rating.merge(ratings_with_df,on='Book-Title')
popular.rename(columns={'Book-Rating':'ratings_counts'},inplace=True)
```

```python
popular= popular[popular['ratings_counts']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python

popular= popular.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','ratings_counts','avg_rating']]
```

<h1> Collaburative Filtering </h1>

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
reader_list=x[x].index
```

```python
filtered_ratings=ratings_with_name[ratings_with_name['User-ID'].isin(reader_list)]
```

```python
filtered_ratings
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(pt)
```

```python
similarity
```

```python
def recommend_books(book_name):
    """Recommends similar books based on a given book name.

    Args:
        book_name (str): The name of the book to find similar recommendations for.

    Returns:
        None
    """

    # Find the index of the book in the dataset
    index = np.where(pt.index == book_name)[0][0]

    # Find the most similar books to the given book
    similar_books = sorted(list(enumerate(similarity[index])), key=lambda x: x[1], reverse=True)[1:6]

    # Print the recommended books
    for book_index, _ in similar_books:  # unpack the tuple (index, similarity)
        print(pt.index[book_index])
```

```python
recommend_books('The Notebook')
```
